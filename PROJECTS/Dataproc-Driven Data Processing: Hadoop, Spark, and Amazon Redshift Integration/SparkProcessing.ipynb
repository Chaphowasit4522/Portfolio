```python
import os
import psutil

# Get CPU information
cpu_info = os.popen("lscpu").read()
print("CPU Information:")
print(cpu_info)

# Get memory information
memory_info = psutil.virtual_memory()
print("\nMemory Information:")
print(f"Total Memory: {memory_info.total / (1024 ** 3):.2f} GB")
print(f"Available Memory: {memory_info.available / (1024 ** 3):.2f} GB")
print(f"Used Memory: {memory_info.used / (1024 ** 3):.2f} GB")
print(f"Memory Usage Percentage: {memory_info.percent}%")
```

```python
import os

disk_info = psutil.disk_usage('/')
print("Disk Space Information:")
print(f"Total Space: {disk_info.total / (1024 ** 3):.2f} GB")
print(f"Used Space: {disk_info.used / (1024 ** 3):.2f} GB")
print(f"Free Space: {disk_info.free / (1024 ** 3):.2f} GB")
print(f"Disk Usage Percentage: {disk_info.percent}%")
```

```python
# create Spark Session for use Spark
from pyspark.sql import SparkSession
spark = SparkSession.builder.master("local[*]").getOrCreate()
```

```python
from pyspark.sql import SparkSession
from pyspark.ml.feature import OneHotEncoder, StringIndexer
from pyspark.ml.classification import RandomForestClassifier, GBTClassifier
from pyspark.ml.evaluation import MulticlassClassificationEvaluator
from pyspark.sql.functions import col
from pyspark.ml import Pipeline


# Load CSV data into a PySpark DataFrame
csv_path = 'gs://dataproc-staging-xxxxxxxxxxxxxxxx/notebooks/jupyter/MCI_2014_to_2017.csv'
df_spark = spark.read.csv(csv_path, header=True, inferSchema=True)
df_spark.show(5)
```

```python
# Convert PySpark DataFrame to Pandas DataFrame
df_pandas = df_spark.toPandas()

# Display the Pandas DataFrame
df_pandas
```

```python
# Select relevant columns
col_list = ['occurrenceyear', 'occurrencemonth', 'occurrenceday', 'occurrencedayofyear',
            'occurrencedayofweek', 'occurrencehour', 'MCI', 'Division', 'Hood_ID', 'premisetype']

# Create a new DataFrame with selected columns
df_selected = df_spark.select(col_list)

# Show the new DataFrame
df_selected.show()
```

```python
# Filter data for occurrences after 2013
df_spark = df_spark.filter(col('occurrenceyear') > 2013)
df_spark.show()
```

```python
# Count the number of rows
num_rows = df_spark.count()

# Get the column names
columns = df_spark.columns

# Count the number of columns
num_columns = len(columns)

# Print the shape
print(f"Shape of the DataFrame: ({num_rows}, {num_columns})")
```

```python
from pyspark.sql.functions import col

# Iterate through columns and check for null values
null_counts = [df_spark.where(col(c).isNull()).count() for c in df_spark.columns]

# Create a dictionary with column names and corresponding null counts
null_counts_dict = dict(zip(df_spark.columns, null_counts))

# Print the null counts for each column
for column, null_count in null_counts_dict.items():
    print(f"Null count in {column}: {null_count}")
```

```python
# Export CSV file
output_path = "gs://dataproc-staging-xxxxxxxxxxxxxxxxxxxxxxxxxx/notebooks/jupyter/"
output_file_name = "CrimeDetection.csv"

df_spark.write.option("header", "true").csv(output_path + output_file_name)
```
